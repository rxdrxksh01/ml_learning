```python
import pandas as pd
import requests
from bs4 import  BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
webpage = requests.get('https://quotes.toscrape.com/page/2/', headers=headers).text
# print(response.status_code)
# print(response.url)
# print("Page Not Found" in response.text)



```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
divs = soup.find_all('div', class_='quote')
# span = div.find_all('span', class_='text')
# print(divs)
qoutes = []
authors = []
tags_list = []

for div in divs:
    span = div.find('span', class_='text')
    diva = div.find('div',class_ = 'tags')
    tags = diva.find_all('a',class_ = 'tag')
    tag_list = [tag.text.strip() for tag in tags]
    tags_list.append(tag_list)

    small = div.find('small', class_ = 'author')
    qoutes.append(
         span.text.strip()

    )
    authors.append(
        small.text.strip())


    # print(span.text.strip())
# print(tags_list)
    # print(small.text.strip())
    # # print(tag)

```

```python
final = pd.DataFrame()
for j in range(1,11):
    url = 'https://quotes.toscrape.com/page/{}/'.format(j)
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    divs = soup.find_all('div', class_='quote')
    # span = div.find_all('span', class_='text')
    # print(divs)
    qoutes = []
    authors = []
    tags_list = []

    for div in divs:
        span = div.find('span', class_='text')
        diva = div.find('div',class_ = 'tags')
        tags = diva.find_all('a',class_ = 'tag')
        tag_list = [tag.text.strip() for tag in tags]
        tags_list.append(tag_list)

        small = div.find('small', class_ = 'author')
        qoutes.append(
            span.text.strip()

        )
        authors.append(
            small.text.strip())
    d = {'quote':qoutes,'author':authors,'tags':tags_list}
    df = pd.DataFrame(d)
    # df.shape
    final = pd.concat([final, df], ignore_index=True)


    

    
    

    
    
```
